```python
# Student Marks Analysis Project

import pandas as pd
import matplotlib.pyplot as plt

# Creating dataset

data = {
    "Name": ["Arun", "Priya", "Rahul", "Sneha", "Karthik"],
    "Subject": ["Math", "Science", "English", "Math", "Science"],
    "Marks": [85, 92, 76, 88, 67],
    "Attendance": [95, 98, 85, 90, 80]
}

df = pd.DataFrame(data)

print("Student Data")
print(df)

# Average Marks
average_marks = df["Marks"].mean()
print("\nAverage Marks:", average_marks)

# Highest Scorer
highest = df.loc[df["Marks"].idxmax()]
print("\nHighest Scorer:")
print(highest)

# Lowest Scorer
lowest = df.loc[df["Marks"].idxmin()]
print("\nLowest Scorer:")
print(lowest)

# Students scoring above 80
above_80 = df[df["Marks"] > 80]
print("\nStudents Scoring Above 80:")
print(above_80)

# Bar Chart

plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Marks"])
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
